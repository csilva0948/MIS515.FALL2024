III. Data Cleaning:

```python
# Check for missing values
data.isnull().sum()
```

```python
# Drow rows with missing target variable
data_cleaned = data.dropna(subset=['Life expectancy '])
```

```python
# Replace missing values for numeric columns with median
num_col = data_cleaned.select_dtypes(include=['float64', 'int64']).columns
data_cleaned.loc[:, num_col] = data_cleaned[num_col].fillna(data_cleaned[num_col].median())
```

```python
# Check for missing values after data cleaning
data_cleaned.isnull().sum()
```

IV. Exploratory Data Analysis (EDA) post-cleaning

```python
# Distribution of Life Expectancy Frequency
plt.figure(figsize=(8,6))
sns.histplot(data_cleaned['Life expectancy '], kde=True)
plt.title('Distribution of Life Expectancy')
plt.xlabel('Life Expectancy')
plt.ylabel('Frequency')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(12,10))
correlation_matrix = data_cleaned.select_dtypes(include=['float64', 'int64']).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Scatter plots for moderate and strong correlation factors that affect life expectancy
features = ['Adult Mortality', 'GDP', 'Schooling', 'Income composition of resources', ' HIV/AIDS', ' BMI ', 'Polio', 'Diphtheria ', ' thinness  1-19 years']
for feature in features:
  plt.figure(figsize=(8,6))
  sns.scatterplot(x=data_cleaned[feature], y=data_cleaned['Life expectancy '])
  plt.title(f'Life Expectancy vs {feature}')
  plt.xlabel(feature)
  plt.ylabel('Life Expectancy')
  plt.show()
```

V. Predictive Modeling

```python
# x, y arrays
x = data_cleaned[features]
y = data_cleaned['Life expectancy ']
```

```python
# Standardisation of the data using StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
# Split the data into an 80/20 training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

1. Linear Regression

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# Model Evaluation
# print the intercept
print(model.intercept_)
```

```python
coeff_data=pd.DataFrame(model.coef_,x.columns,columns=['Coefficient'])
coeff_data
```

```python
# Preditions
pred_linear = model.predict(x_test)

# Actual vs Predicted for Linear Regression
plt.figure(figsize=(8, 6))
plt.scatter(y_test, pred_linear, alpha=0.7)
plt.title('Actual vs Predicted Life Expectancy (Linear Regression)')
plt.xlabel('Actual Life Expectancy')
plt.ylabel('Predicted Life Expectancy')
plt.axline((0, 0), slope=1, color='purple', linestyle='-', label='Ideal Fit')
plt.legend()
plt.show()
```

```python
# Evaluate Linear Regression Metrics
from sklearn import metrics
r2_linear = metrics.r2_score(y_test,pred_linear)

print('Mean Absolute Error (MAE): ', metrics.mean_absolute_error(y_test,pred_linear))
print('Mean Squared Error (MSE): ', metrics.mean_squared_error(y_test,pred_linear))
print('RMSE: ', np.sqrt(metrics.mean_squared_error(y_test, pred_linear)))
print('R2 Score: ', metrics.r2_score(y_test,pred_linear))
```

2. Random Forest Regressor

```python
# Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor(n_estimators=500, criterion='squared_error', max_features= 3, bootstrap=True, random_state=1234)
rf_model.fit(x_train, y_train)
```

```python
# Predictions
pred_rf = rf_model.predict(x_test)

# Actual vs Predicted for Random Forest Regressor
plt.figure(figsize=(8, 6))
plt.scatter(y_test, pred_rf, alpha=0.7)
plt.title('Actual vs Predicted Life Expectancy (Random Forest Regressor)')
plt.xlabel('Actual Life Expectancy')
plt.ylabel('Predicted Life Expectancy')
plt.axline((0, 0), slope=1, color='green', linestyle='-', label='Ideal Fit')
plt.legend()
plt.show()
```

```python
# Evaluate Random Forest Regressor Metrics
from sklearn import metrics
print('Mean Absolute Error (MAE): ', metrics.mean_absolute_error(y_test,pred_rf))
print('Mean Squared Error (MSE): ', metrics.mean_squared_error(y_test,pred_rf))
print('RMSE: ', np.sqrt(metrics.mean_squared_error(y_test, pred_rf)))
print('R2 Score: ', metrics.r2_score(y_test, pred_rf))
```
